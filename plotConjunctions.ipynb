```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
gr = pd.read_csv('../tables/conjunctionsGR.csv')
gf = pd.read_csv('../tables/conjunctionsGF.csv')
```

```python
gr['radar_group'] = gr['radar']

for i in iter(range(0,len(gr))):

    radar = gr['radar'][i]

    if str(radar)[0:3] == 'Mil':
        gr['radar_group'][i] = 'Millstone Hill'
    elif str(radar[0:3]) == 'Are':
        gr['radar_group'][i] = 'Arecibo'
    elif str(radar[0:3]) == 'EIS':
        gr['radar_group'][i] = 'EISCAT'
    elif str(radar[0:3]) == 'Jic':
        gr['radar_group'][i] = 'Jicamarca'
    elif str(radar[0:3]) == 'Kha':
        gr['radar_group'][i] = 'Kharkov'
    elif str(radar[0:3]) == 'MU ':
        gr['radar_group'][i] = 'MU'
    elif str(radar[0:3]) == 'Res':
        gr['radar_group'][i] = 'Resolute Bay'
    else:
        pass
```

```python
gf['radar_group'] = gf['radar']

for i in iter(range(0,len(gf))):

    radar = gf['radar'][i]

    if str(radar)[0:3] == 'Mil':
        gf['radar_group'][i] = 'Millstone Hill'
    elif str(radar[0:3]) == 'Are':
        gf['radar_group'][i] = 'Arecibo'
    elif str(radar[0:3]) == 'EIS':
        gf['radar_group'][i] = 'EISCAT'
    elif str(radar[0:3]) == 'Jic':
        gf['radar_group'][i] = 'Jicamarca'
    elif str(radar[0:3]) == 'Kha':
        gf['radar_group'][i] = 'Kharkov'
    elif str(radar[0:3]) == 'MU ':
        gf['radar_group'][i] = 'MU'
    elif str(radar[0:3]) == 'Res':
        gf['radar_group'][i] = 'Resolute Bay'
```

```python
df = gr
for radar_name in df.radar_group.unique():
    fig, ax = plt.subplots(figsize=(10, 5))
    plt.title('GRACE - {radar}'.format(radar = radar_name))

    df_plot = df[df.radar_group == radar_name]
    plt.scatter(df_plot['date'], df_plot['ne_diff'])


    plt.show()
    # plt.savefig("../figures/diff_grace_{name}.png".format(name = radar_name))
    plt.close()
```

```python
for radar_name in df.radar_group.unique():
    fig, ax = plt.subplots(figsize=(7.5, 5))
    plt.title('GRACE - {radar}'.format(radar = radar_name))

    df_plot = df[df.radar_group == radar_name]

    try:

        plt.hist(df_plot['ne_diff'], bins=20)

    except:
        plt.hist(df_plot['ne_diff'])



    plt.xlabel('Difference GRACE Absolute Ne - Radar Ne')
    plt.ylabel('Count')

    # plt.show()
    plt.savefig("../figures/diff_grace_{name}.png".format(name = radar_name))
    plt.close()
```

```python
gr = pd.read_csv('../tables/GR_conjunctions_clean.csv')
gr["radar_mnemonic_unique"] = gr["radar_mnemonic"].str[0:3]

gr['location'] = gr['radar_mnemonic_unique']

for i in iter(range(0, len(gr))):

    radar = gr['radar_mnemonic_unique'][i]

    if str(radar)[0:3] == 'mlh':
        gr['location'][i] = 'Millstone Hill'
    elif str(radar[0:3]) == 'arg':
        gr['location'][i] = 'Arecibo'
    elif str(radar[0:3]) == 'arl':
        gr['location'][i] = 'Arecibo'
    elif str(radar[0:3]) == 'eis':
        gr['location'][i] = 'Trons'
    elif str(radar[0:3]) == 'tro':
        gr['location'][i] = 'Trons'
    elif str(radar[0:3]) == 'jro':
        gr['location'][i] = 'Jicamarca'
    elif str(radar[0:3]) == 'kpi':
        gr['location'][i] = 'Kharkov'
    elif str(radar[0:3]) == 'mui':
        gr['location'][i] = 'MU'
    elif str(radar[0:3]) == 'ran':
        gr['location'][i] = 'Resolute Bay'
    elif str(radar[0:3]) == 'ras':
        gr['location'][i] = 'Resolute Bay'
    elif str(radar[0:3]) == 'lyr':
        gr['location'][i] = 'Svalbard'
    elif str(radar[0:3]) == 'pfa':
        gr['location'][i] = 'Poker Flat'

    gr = gr[~gr.location.str.contains("Kharkov", na=False)]
    gr = gr[~gr.location.str.contains("MU", na=False)]
```
