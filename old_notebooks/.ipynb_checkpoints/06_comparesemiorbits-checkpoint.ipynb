```python
# import libraries
import os
import pandas as pd
import datetime
from matplotlib import pyplot as plt
import numpy as np
import glob
import datetime
from scipy.signal import argrelextrema
from scipy.signal import find_peaks
import math
from tqdm import tqdm
from scipy import interpolate

```

```python
# auxiliary functions

# add datetime index
def addDatetimes(df,dftype):
    datetimes = [] 
    
    if dftype == 'mlh':
        for i in iter(range(0,len(df))):
            datetimes.append(datetime.datetime(df['YEAR'][i],
                                               df['MONTH'][i],
                                               df['DAY'][i],
                                               df['HOUR'][i],
                                               df['MIN'][i],
                                               df['SEC'][i]) )
    if dftype == 'grf':
        for i in iter(range(0,len(df))):
            datetimes.append(datetime.datetime(df['yy'][i],
                                               df['mo'][i],
                                               df['dd'][i],
                                               df['hh'][i],
                                               df['mm'][i],
                                               df['ss'][i]) )


        
    df['datetimes'] = datetimes
    df.index = df.datetimes
    
    return df

# define the true objective function
def objective(x, a, b):
    return  a*np.exp(b*x)

# find nearest value in a array 
def find_nearest_idx(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return idx

# estimate earth radius 
def earth_radius (B):
    B=math.radians(B) #converting into radians
    a = 6378.137  #Radius at sea level at equator
    b = 6356.752  #Radius at poles
    c = (a**2*math.cos(B))**2
    d = (b**2*math.sin(B))**2
    e = (a*math.cos(B))**2
    f = (b*math.sin(B))**2
    R = math.sqrt((c+d)/(e+f))
    return R

# estimate local time zone
def calc_solartime(longitude_local, datetime_UTC):

    longitudes = []
    for i in iter(range(0,13)):
        longitudes.append(7.5 + (15*i))    
    
    longitudes.append(np.abs(longitude_local))
    longitudes.sort()
    
    idx_localtime = np.argwhere(longitudes == np.abs(longitude_local))[0][0]
    if longitude_local >=0:
        datetime_local = datetime_UTC + datetime.timedelta(hours = int(idx_localtime))
    else:
        datetime_local = datetime_UTC - datetime.timedelta(hours = int(idx_localtime))
    
    return datetime_local
```

```python
grf = pd.read_csv('all_grf.csv') # file file contains altitude estimated by latitude and localtime
grf_jan = pd.read_csv('all_grf_jan.csv') # file file contains altitude estimated by latitude and localtime
```

```python
print(len(grf))
print(len(grf_jan))
```

```python
# convert dates to datetime
grf.datetimes = pd.to_datetime(grf.datetimes)
grf.datetimes_local = pd.to_datetime(grf.datetimes_local)

grf_jan.datetimes = pd.to_datetime(grf_jan.datetimes)
grf_jan.datetimes_local = pd.to_datetime(grf_jan.datetimes_local)
```

```python
# extract hour
grf['hour'] = grf.datetimes_local.apply(lambda x : x.hour)

grf_jan['hour'] = grf_jan.datetimes_local.apply(lambda x : x.hour)
```

```python
plt.rcParams.update({'font.size': 18})

from scipy.signal import argrelextrema


orbits_idx = find_peaks(np.abs(grf.qd_lat.values))[0]

idx_i = []
idx_j = []
idx_eq0 = []
list_hour = []

for i in tqdm(range(1,len(orbits_idx))):
    try:

        idx_i_value = orbits_idx[i-1]
        idx_j_value = orbits_idx[i]

        idx_i.append(idx_i_value)
        idx_j.append(idx_j_value)

        idx_eq0_value = np.abs(grf['qd_lat'][idx_i_value:idx_j_value].values).argmin() + idx_i_value
        idx_eq0.append(idx_eq0_value)

        list_hour_value = grf.hour[idx_eq0_value]
        list_hour.append(list_hour_value)

    except:
        print(idx_i_value,idx_j_value)

        
df_idx = pd.DataFrame(
    {'list_hour': list_hour,
     'idx_eq0': idx_eq0,
     'idx_i' : idx_i,
     'idx_j': idx_j
    })
```

```python
plt.rcParams.update({'font.size': 18})

from scipy.signal import argrelextrema


orbits_idx = find_peaks(np.abs(grf_jan.qd_lat.values))[0]

idx_i = []
idx_j = []
idx_eq0 = []
list_hour = []

for i in tqdm(range(1,len(orbits_idx))):
    try:

        idx_i_value = orbits_idx[i-1]
        idx_j_value = orbits_idx[i]

        idx_i.append(idx_i_value)
        idx_j.append(idx_j_value)

        idx_eq0_value = np.abs(grf_jan['qd_lat'][idx_i_value:idx_j_value].values).argmin() + idx_i_value
        idx_eq0.append(idx_eq0_value)

        list_hour_value = grf_jan.hour[idx_eq0_value]
        list_hour.append(list_hour_value)

    except:
        print(idx_i_value,idx_j_value)

        
df_idx_jan = pd.DataFrame(
    {'list_hour': list_hour,
     'idx_eq0': idx_eq0,
     'idx_i' : idx_i,
     'idx_j': idx_j
    })
```

```python
print(len(df_idx))
print(len(df_idx_jan))
```

```python
pd.options.mode.chained_assignment = None
import matplotlib.ticker as ticker


# loop through hours
for i in tqdm(range(0,24)):
    # select hour
    df_plot = df_idx[df_idx['list_hour'] == i]
    df_plot = df_plot.reset_index()
    
    df_plot_jan = df_idx_jan[df_idx_jan['list_hour'] == i]
    df_plot_jan = df_plot_jan.reset_index()
    
    fig = plt.figure(figsize=(20,4))
    plt.suptitle('Hour {i}'.format(i=i),fontweight='bold')
    
    ax = plt.subplot(1,3,1)
    hour_segments = []
    for j in iter(range(0,len(df_plot))):
        
        idx_i = df_plot['idx_i'][j]
        idx_j = df_plot['idx_j'][j]
        
        grf_plot = grf[idx_i:idx_j]
        hour_segments.append(grf_plot)
        plt.plot(grf_plot['qd_lat'], grf_plot['ne'],'r')
        
    hour0 = pd.concat(hour_segments)

    plt.xlim(-90,90)
    ax.xaxis.set_major_locator(ticker.MultipleLocator(30))
    plt.xlabel('magnetic latitude')
    plt.ylabel('electron density')
    
    hour_segments2 = []
    for k in iter(range(0,len(hour_segments))):
    #     print(k)
        df = hour_segments[k]

        x= df['qd_lat'].values
        y = df['ne'].values

        f = interpolate.interp1d(x, y, fill_value="extrapolate")

        xnew = np.arange(int(x.min()), int(x.max()), 1)
        ynew = f(xnew)

        df2 = pd.DataFrame(
            {'qd_lat': xnew,
             'ne': ynew
            })
        hour_segments2.append(df2)
        
    hour1 = pd.concat(hour_segments2)
    
    ax = plt.subplot(1,3,2)
    hour_segments_jan = []
    for j in iter(range(0,len(df_plot_jan))):
        
        idx_i = df_plot_jan['idx_i'][j]
        idx_j = df_plot_jan['idx_j'][j]
        
        grf_plot_jan = grf_jan[idx_i:idx_j]
        hour_segments_jan.append(grf_plot_jan)
        plt.plot(grf_plot_jan['qd_lat'], grf_plot_jan['ne'],'k')
        
    hour0_jan = pd.concat(hour_segments_jan)

    plt.xlim(-90,90)
    ax.xaxis.set_major_locator(ticker.MultipleLocator(30))
    plt.xlabel('magnetic latitude')
    plt.ylabel('electron density')
    
    hour_segments2_jan = []
    for k in iter(range(0,len(hour_segments_jan))):
    #     print(k)
        df_jan = hour_segments_jan[k]

        x= df_jan['qd_lat'].values
        y = df_jan['ne'].values

        f = interpolate.interp1d(x, y, fill_value="extrapolate")

        xnew = np.arange(int(x.min()), int(x.max()), 1)
        ynew = f(xnew)

        df2_jan = pd.DataFrame(
            {'qd_lat': xnew,
             'ne': ynew
            })
        hour_segments2_jan.append(df2_jan)
        
    hour1_jan = pd.concat(hour_segments2_jan)
    
    
    

    # extract mean, max, min
    grf_groupby = hour1.groupby('qd_lat').agg({'ne': ['mean', 'min', 'max']})
    grf_groupby_jan = hour1_jan.groupby('qd_lat').agg({'ne': ['mean', 'min', 'max']})

        
    ax = plt.subplot(1,3,3) 
    plt.plot(grf_groupby.index, grf_groupby['ne']['mean'],'r', label ='Chao')
    plt.plot(grf_groupby_jan.index, grf_groupby_jan['ne']['mean'],'k', label = 'Jan')
    plt.xlim(-90,90)
    ax.xaxis.set_major_locator(ticker.MultipleLocator(30))
    plt.xlabel('magnetic latitude')
    plt.ylabel('electron density')
    plt.legend()


#     ax = plt.subplot(1,3,3) 
#     plt.plot(grf_groupby.index, grf_groupby['ne']['mean'])
#     plt.plot(grf_groupby.index, grf_groupby['ne']['max'])     
#     plt.plot(grf_groupby.index, grf_groupby['ne']['min'])     
#     plt.xlim(-90,90)
#     ax.xaxis.set_major_locator(ticker.MultipleLocator(30))
#     plt.xlabel('magnetic latitude')
#     plt.ylabel('electron density')
```
