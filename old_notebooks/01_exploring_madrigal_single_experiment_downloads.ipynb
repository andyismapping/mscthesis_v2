```python
# import libraries
import pandas as pd
import datetime
from matplotlib import pyplot as plt
import numpy as np
```

```python
# load datasets
jro = pd.read_csv('../data/external/Madrigal/01_exploring_madrigal/jro20200114a.001.txt', sep = '\s+')

aro = pd.read_csv('../data/external/Madrigal/01_exploring_madrigal/aro20190205clpg.003.txt', sep = '\s+')

mlh_i = pd.read_csv('../data/external/Madrigal/01_exploring_madrigal/mlh200130i.004.txt', sep = '\s+')
mlh_j = pd.read_csv('../data/external/Madrigal/01_exploring_madrigal/mlh200130j.004.txt', sep = '\s+')
mlh_k = pd.read_csv('../data/external/Madrigal/01_exploring_madrigal/mlh200130k.004.txt', sep = '\s+')
mlh_l = pd.read_csv('../data/external/Madrigal/01_exploring_madrigal/mlh200130l.004.txt', sep = '\s+')
mlh_m = pd.read_csv('../data/external/Madrigal/01_exploring_madrigal/mlh200130m.002.txt', sep = '\s+')
mlh_g = pd.read_csv('../data/external/Madrigal/01_exploring_madrigal/mlh200130g.003.txt', sep = '\s+')

tro_uhf = pd.read_csv('../data/external/Madrigal/01_exploring_madrigal/MAD6400_2019-01-03_beata_V223@uhf.txt', sep = '\s+')
tro_vhf = pd.read_csv('../data/external/Madrigal/01_exploring_madrigal/MAD6200_2020-03-25_gup_31.3@vhf.txt', sep = '\s+')

sod_uhf = pd.read_csv('../data/external/Madrigal/01_exploring_madrigal/realtime.asc.txt', sep = '\s+')
```

```python
# add datetimes
def addDatetimes(df):
    datetimes = [] 
    for i in iter(range(0,len(df))):
        datetimes.append(datetime.datetime(df['YEAR'][i],
                                           df['MONTH'][i],
                                           df['DAY'][i],
                                           df['HOUR'][i],
                                           df['MIN'][i],
                                           df['SEC'][i]) )
        
    df['datetimes'] = datetimes
    df.index = df.datetimes
    
    return df

jro = addDatetimes(jro)

aro = addDatetimes(aro)

mlh_i = addDatetimes(mlh_i)
mlh_j = addDatetimes(mlh_j)
mlh_k = addDatetimes(mlh_k)
mlh_l = addDatetimes(mlh_l)
mlh_m = addDatetimes(mlh_m)
mlh_g = addDatetimes(mlh_g)

tro_uhf = addDatetimes(tro_uhf)
tro_vhf = addDatetimes(tro_vhf)

sod_uhf = addDatetimes(sod_uhf)
```
