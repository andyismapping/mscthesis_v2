```python
import pandas as pd
instVariables = pd.read_csv('../data/external/Madrigal/Metadata/madrigalInstMetadata.csv')
# longitude = (((instVariables['Longitude (-180-180)'] + 180) % 360) - 180)
# longitude
instVariables['Longitude (-180-180)']
```

```python
import numpy as np
longitudes = []

for i in iter(range(0,len(instVariables))):
    try:
        longitudes.append(((float(instVariables['Longitude (-180-180)'][i]) + 180) % 360) - 180)
    except:
        longitudes.append(np.nan)
```

```python
latitudes = list(instVariables['Latitude'])
names = list(instVariables['Name'])
```

```python
radars = pd.DataFrame({'name':names,
              'latitude': latitudes,
              'longitude': longitudes})
radars.to_csv('../radars.csv')
           
```
