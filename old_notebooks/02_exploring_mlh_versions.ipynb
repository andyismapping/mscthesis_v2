```python
# import libraries
import pandas as pd
import datetime
from matplotlib import pyplot as plt
import numpy as np
```

```python
# load datasets ##check ssd for the data, its not where its supposed to be
mlh_v2 = pd.read_csv('../data/external/Madrigal/02_exploring_mlh_versions/mlh180607i.002_10001345.txt', sep = '\s+')
mlh_v4 = pd.read_csv('../data/external/Madrigal/02_exploring_mlh_versions/mlh180607i.004_10001345.txt', sep = '\s+')


# add datetime index
def addDatetimes(mlh_df):
    datetimes = [] 
    for i in iter(range(0,len(mlh_df))):
        datetimes.append(datetime.datetime(mlh_df['YEAR'][i],
                                           mlh_df['MONTH'][i],
                                           mlh_df['DAY'][i],
                                           mlh_df['HOUR'][i],
                                           mlh_df['MIN'][i],
                                           mlh_df['SEC'][i]) )
        
    mlh_df['datetimes'] = datetimes
    mlh_df.index = mlh_df.datetimes
    
    return mlh_df

mlh_v2 = addDatetimes(mlh_v2)
mlh_v4 = addDatetimes(mlh_v4)

# compare outputs
print(np.mean(mlh_v2-mlh_v4))

# get the date we know has 2 measurments
date = mlh_v2.index[0]

# filter datasets

mlh_v2_plot = mlh_v2[mlh_v2.index == date]
mlh_v4_plot = mlh_v4[mlh_v4.index == date]


# final plot
plt.rcParams.update({'font.size': 18})
fig, ax = plt.subplots(figsize=(10,15))
plt.plot(np.log10(mlh_v2_plot.NE), mlh_v2_plot.GDALT, label = 'exp 10001345 v002',linestyle='--', marker='o')
plt.plot(np.log10(mlh_v4_plot.NE), mlh_v4_plot.GDALT, label = 'exp 10001345 v004',linestyle='--', marker='o')



plt.legend()

plt.ylabel('altitude in km')
plt.xlabel('log of Ne')


plt.title('Millstone Hill - Zenith single-pulse \n {date}'.format(date=date))
```
